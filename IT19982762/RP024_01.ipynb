```python
from ultralytics import YOLO
```

```python
model = YOLO("yolo11m.pt")
```

```python
model.train(data = "/content/drive/MyDrive/RP024-01/data_paths.yaml", imgsz= 640, batch =8 , epochs =100, workers=1 , device=0, patience=10)
```

```python
best_model = YOLO("/content/drive/MyDrive/RP024-01/runs/detect/train/weights/best.pt")
```

```python
results = best_model.predict("/content/drive/MyDrive/RP024-01/val/images/068b29b1-15_coarsedropout.jpg", show=False, conf=0.5, save=False)
results[0].show()
```
